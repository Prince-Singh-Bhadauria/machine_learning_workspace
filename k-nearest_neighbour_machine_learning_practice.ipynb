# Training & Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, random_state = 0)

#random_state will shuffle data before dividing (in 75% and 25%)
```

```python
print('X_train shape: {}'.format(X_train.shape))
print('y_train shape: {}'.format(y_train.shape))
```

```python
print('X_test shape: {}'.format(X_test.shape))
print('y_test shape: {}'.format(y_test.shape))
```

# Building Model : k-Nearest Neighbours

```python
# importing the KNeighboursClassifier, class from sk learn, its like a clay for the model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# fitting our data to train the model, its like giving desired shape to the clay
knn.fit(X_train, y_train)
```

# Making Predictions

```python
X_new = np.array([[5,2.9,1,0.2]])
print('X_new.shape : \n {}'.format(X_new.shape))
# note: we have made the measurements of this single flower into a row
```

```python
# To make predictions we all the 'predict()' method of the 'knn' object
prediction = knn.predict(X_new)
print('Prediction: {}'.format(prediction))
print('Predicted target name: {}'. format(iris_dataset['target_names'][prediction]))
```

# Evaluating the Model

we can make a prediction for each iris in the test data and compare it against its label(the known species).

we can measure how well the model works by computing the ACCURACY

```python
y_pred = knn.predict(X_test)
print('Test set predictions: \n {}'.format(y_pred))
```

```python
# Computing the test set's accuracy
print ('Test set score: {:.2f}'.format(np.mean(y_pred == y_test)))
```

```python
# computing accuracy using SCORE() method
print('The set score: {}'.format(knn.score(X_test, y_test)))
```
